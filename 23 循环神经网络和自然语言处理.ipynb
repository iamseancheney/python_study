# 循环神经网络

## RNN的结构

左图，一层只有一个神经元：每一轮的输出会成为下一轮的输入。
右图是一个神经元随时间展开的状态。

![](https://tva1.sinaimg.cn/large/006y8mN6ly1g94b0dksi9j30de05uwfe.jpg)

一层只有一个神经元的话，拟合能力不强。
所以变成，一层有多个神经元。

![](https://tva1.sinaimg.cn/large/006y8mN6ly1g94b0v5wgaj30de053q43.jpg)

单个矢量 x<sub>(t)</sub> 输入，时刻t的输出y：

![](https://tva1.sinaimg.cn/large/006y8mN6ly1g94b15ckpyj30c001et8o.jpg)

一次只计算单个矢量，速度太慢，所以要变成一次计算一组矢量。一个批次的矢量构成了矩阵 X<sub>(t)</sub> ：

![](https://tva1.sinaimg.cn/large/006y8mN6ly1g94b1kcjldj30de02zwen.jpg)

其中，

- Y<sub>(t)</sub> 是时刻t的输出，m行，n<sub>neurons</sub>列的矩阵，m是批次的中实例数，n<sub>neurons</sub>是神经元数；
- X<sub>(t)</sub> 是时刻t的输入，m行，n<sub>inputs</sub>列的矩阵，n<sub>inputs</sub>是输入的特征数；
- W<sub>x</sub> 是当前时刻t的输入连接权重矩阵，n<sub>inputs</sub>行，n<sub>neurons</sub>列；
- W<sub>y</sub> 是上一时刻t-1的输出连接权重矩阵，n<sub>neurons</sub>行，n<sub>neurons</sub>列；
- b是每个神经元的偏置矢量，大小为n<sub>neurons</sub>；
- W是W<sub>x</sub>和W<sub>y</sub>的垂直连接矩阵，形状是 (n<sub>inputs</sub> + n<sub>neurons</sub>) 行，n<sub>neurons</sub>列；
- [X<sub>(t)</sub> Y<sub>(t-1)</sub>] 表示矩阵X<sub>(t)</sub>和Y<sub>(t-1)</sub>的水平连接。

> **神经元的记忆性**
> 
> 时刻t的输出是时刻t的输入和时刻t-1的输出的函数，后者又是时刻t-1的输入和时刻t-2的输出的函数，以此类推。对于第一个时刻t=0，因为没有上一步输出，对应的矩阵都是0。

## 不同类型的RNN

![](https://tva1.sinaimg.cn/large/006y8mN6ly1g9a9dao2wjj30j40cbjut.jpg)

左上：序列到序列，sequence-to-sequence，股票预测：输入过去N天的数据，输出过去N-1天加明天的数据；

右上：序列到矢量，sequence-to-vector，除了最后一个输出，忽略所有，情感分析：输入一段sequence-to-vector，解码器是
左下：矢量到序列，vector-to-sequence，将同样的矢量在每个时间步骤输入，输出一个序列，图像生成标题：输入图片，得到和图片内容有关的标题；

右下：编码器+解码器，编码器是sequence-to-vector，解码器是vector-to-sequence，翻译：先将一段话转化为矢量，再用解码器将矢量变为另一种语言。

> 编码器+解码器由于sequence-to-sequence的原因：一句话的后面的词会影响到整句话的意思（比如日语总是在一句话的末尾表示否定），就像口译中有：交替传译vs同声传译（交替传译的效果由于同声传译，但是间隔长）
