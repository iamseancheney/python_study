```python
import pandas as pd
import numpy as np
from numpy import nan as NA

data = pd.Series([1, NA, 3.5, NA, 7])
data
```

```python
data.dropna()
```

```python
# 对于DataFrame对象。dropna默认丢弃任何含有缺失值的行：

data = pd.DataFrame([[1., 6.5, 3.], [1., NA, NA],
                     [NA, NA, NA], [NA, 6.5, 3.]])
data
```

```python
cleaned = data.dropna()
cleaned
```

```python
# 传入how='all'将只丢弃全为NA的那些行：

data.dropna(how='all')
```

```python
# 用这种方式丢弃列，只需传入axis=1即可：

data[4] = NA
data
```

```python
data.dropna(axis=1, how='all')
```

```python
# thresh参数

df = pd.DataFrame(np.random.randn(7, 3))
df.iloc[:4, 1] = NA
df.iloc[:2, 2] = NA
df
```

```python
df.dropna()
```

```python
df.dropna(thresh=2)
```

```python
# 填充缺失数据

df.fillna(0)
```

```python
# 通过一个字典调用fillna，就可以实现对不同的列填充不同的值：

df.fillna({1: 0.5, 2: 0})
```

```python
# ffill

df = pd.DataFrame(np.random.randn(6, 3))
df.iloc[2:, 1] = NA
df.iloc[4:, 2] = NA
df
```

```python
df.fillna(method='ffill')
```

```python
# 删除重复数据

data = pd.DataFrame({'k1': ['one', 'two'] * 3 + ['two'],
                         'k2': [1, 1, 2, 3, 3, 4, 4]})
data
```

```python
# 用duplicated()查看是否有重复项

data.duplicated()
```

```python
# drop_duplicates()删除重复项

data.drop_duplicates()
```

```python
# 只希望通过某一列删除重复项

data['v1'] = range(7)
data.drop_duplicates(['k1'])
```

```python
# 默认保留的是第一个出现的值组合。传入keep='last'则保留最后一个：

data.drop_duplicates(['k1', 'k2'], keep='last')
```
