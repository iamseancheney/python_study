```python
import requests

data = {
    'name': 'geremy',
    'age': 22
}
r_2 = requests.get("http://httpbin.org/get", params=data)
print(r_2.text)
```

```python
# 添加请求头

import requests

headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36'
}

r = requests.get("https://www.zhihu.com/explore", headers=headers)

print(r.text)
```

```python
# 二进制文件：图片和视频

import requests

r = requests.get("https://github.com/favicon.ico")
with open('favicon.ico', 'wb') as f:
    f.write(r.content)
```

```python
# post请求

import requests

data = {'name': 'germey', 'age': '22'}
r = requests.post("http://httpbin.org/post", data=data)
print(r.text)
```

```python
# 超时

import requests

response = requests.get('http://github.com', timeout=0.1)

print(response.text)
```

```python
import requests

s = requests.Session()
s.get('http://httpbin.org/cookies/set/sessioncookie/123456789')
r = s.get("http://httpbin.org/cookies")
print(r.text)
```

```python
# 使用代理

import requests

proxies = {
  "http": "49.83.162.13:808",
  "https": "49.83.162.13:808"
}
r = requests.get("https://www.baidu.com", proxies=proxies)
print(r.text)
```

```python
# 处理编码 —— 给大家两段万能代码

import requests
url = 'https://www.baidu.com'

res = requests.get(url).text
try:
    res = res.encode('ISO-8859-1').decode('utf-8')
except:
    try:
        res = res.encode('ISO-8859-1').decode('gbk')
    except:
        res = res
print(res)
```

```python
# -*- coding: utf-8 -*-

import requests

url = 'https://www.baidu.com'
response = requests.get(url)

print(response.encoding)

if response.encoding != 'ISO-8859-1':
    html = response.text
else:
    html = response.content
    if 'charset' not in response.headers.get('content-type'):
        encodings = requests.utils.get_encodings_from_content(response.text)
        if len(encodings) > 0:
            response.encoding = encodings[0]
            html = response.text
            
print(html)
```
