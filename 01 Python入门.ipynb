## 时间线

生日：1956年1月31日

教育背景：阿姆斯特丹大学数学和计算机硕士

1989年，任职于荷兰数学和计算机科学研究学会（相当于中科院数学和计算机所？）的范罗瑟姆在圣诞节期间开始着手开发Python。

路径演变：SETL > ABC > Python

1991 年，van Rossum 通过 alt.sources 新闻组向全世界发布了 Python。

Van Rossum 在1999年的一份报告中，描述了Python的目标：

- It should be an easy and intuitive language, just as powerful as major competitors.
- 简单、直观、强大

- It should be open source, so anyone can contribute to its development.
- 开源

- Its code should be understandable as plain English.
- 容易理解的英语

- It should be suitable for everyday tasks, allowing for short development times.
- 适用于日常工作，开发快速

目前，Python基本是一年一个小版本的更新速度，现在是Python3.8（2019年10月14日）

## 为什么是Python？

- 语法简单流畅，在众多编程语言中最接近人类语言

- 内置库和第三方库众多，可以“开箱即用”（这两点可以让Python开发特别快）

- 一专多能（Python用途非常广泛：脚本、后端、爬虫、数据分析、人工智能）

- 赶上了好时候（大数据 + 人工智能，当代软件开发具有内敛型，即内部复杂，但向外暴露的很简单）

- 边缘计算弥补了Python的计算能力

- 多写（在实践中锻炼总结）
